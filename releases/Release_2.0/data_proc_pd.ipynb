### Колонка адрес

```python
df_moscow.columns = ["ID", "Метро", "Адрес", "Описание", "Балкон", "Санузел"]
df_moscow[df_moscow["Адрес"].isnull()]
```

* Пропусков нет
* Разделяем на улицу и номер дома

```python
def street(string):  # Выделяет из строки адреса только улицу
    str_list = string.split(", ")
    temp = set(
        [
            "улица",
            "переулок",
            "проспект",
            "бульвар",
            "проезд",
            "ш.",
            "ул.",
            "шоссе",
            "пр.",
            "пр-кт",
            "набережная",
        ]
    )
    for i in range(1, len(str_list)):
        x = str_list[i].split()
        for c in x:
            if c in temp:
                return str_list[i]
    return np.nan


def house_num(string):  # Выделяет из строки адреса номер дома
    str_list = string.split(", ")
    temp = set(
        [
            "улица",
            "переулок",
            "проспект",
            "бульвар",
            "проезд",
            "ш.",
            "ул.",
            "шоссе",
            "пр.",
            "пр-кт",
            "набережная",
        ]
    )
    ind = 0
    for i in range(1, len(str_list)):
        x = str_list[i].split()
        for c in x:
            if c in temp:
                ind = i + 1
                break
    # if len(str_list) > ind:
    #     return str_list[ind]
    # else:
    #     return 0
    num = ""
    if 0 < ind < len(str_list):
        for c in str_list[ind]:
            if c.isdigit() == True:
                num += c
            else:
                break
    if num != "":
        return np.int64(num)
    else:
        return 0


df_moscow["Street"] = df_moscow["Адрес"].map(street)
df_moscow["House №"] = df_moscow["Адрес"].map(house_num)

df_moscow["Street"] = df_moscow["Street"].map(
    lambda x: x[: x.find("(") - 1] if type(x) == str and x.find("(") != -1 else x
)
df_moscow["Street"] = df_moscow["Street"].fillna(0)
# df_moscow.loc[::, ["House №", "Street"]].head()
df_moscow[df_moscow["Street"].isna()]
```

### Колонка "Метро"

* Убираем пропуски - ищем в других записях такую же улицу или ставим 0
* Отделяем время до метро 
* В колонке метро только метро
* Колонка "Время до метро пешком"
* Время на машине * 5 = время пешком
* Создаем колонку 'Округ' + категоризуем

#### Отделение названия станции метро и чистка колонки Metro от Nan

```python
def metro_split(string):  # выделяет название станции метро
    if type(string) == str:
        return string[3 : string.find("(") - 1]
    else:
        return string


df_moscow["Metro"] = df_moscow["Метро"].map(metro_split)

df_moscow["Metro"] = df_moscow["Metro"].fillna(
    df_moscow.groupby("Street")["Metro"].transform(
        lambda x: x.mode().iloc[0] if not x.mode().empty else 0
    )
)

# df_moscow[df_moscow["Metro"].isna()]
df_moscow[df_moscow["Metro"] == 0]
```

#### Отделяем время до метро и пересчитываем

* Пропуски заполняем максимальным значением для данной улицы
* При невозможности - медианой дпо всем временам

```python
def metro_time(string):  # выделяет название станции метро
    res = ""
    if type(string) != str:
        return string
    elif string.find("(") == -1:
        return np.nan
    for c in string[string.find("(") + 1 :]:
        if c.isdigit():
            res += c
        else:
            break
    # return res
    if res == "":
        return np.nan
    elif string[string.rfind(" ") + 1 : string.rfind(")")] != "пешком":
        return int(res) * 5
    else:
        return int(res)


df_moscow["Time to metro"] = df_moscow["Метро"].map(metro_time)
#
df_moscow["Time to metro"] = df_moscow["Time to metro"].fillna(
    df_moscow.groupby("Street")["Time to metro"].transform(lambda x: x.max())
)
df_moscow["Time to metro"] = df_moscow["Time to metro"].fillna(
    df_moscow["Time to metro"].median()
)
df_moscow[df_moscow["Time to metro"].isna()]
# df_moscow.dtypes
# df_moscow
```

### Категоризуем готовые колонки

#### Street

```python
unique_streets = df_moscow_1.loc[df_moscow_1["Street"] != "0", "Street"].unique()
street_to_num = {name: i + 1 for i, name in enumerate(unique_streets)} # словарь значений

df_moscow_1["Street_cat"] = (
    df_moscow_1["Street"].map(street_to_num).fillna(0).astype(int)
)
df_moscow_1.dtypes
```

#### Metro

```python
unique_metro = df_moscow_1.loc[df_moscow_1["Metro"] != "0", "Metro"].unique()
metro_to_num = {name: i + 1 for i, name in enumerate(unique_metro)} # словарь значений

df_moscow_1["Metro_cat"] = df_moscow_1["Metro"].map(metro_to_num).fillna(0).astype(int)
df_moscow_1.dtypes
```

#### District

```python
unique_district = df_moscow_1.loc[df_moscow_1["District"] != "0", "District"].unique()
district_to_num = {name: i + 1 for i, name in enumerate(unique_district)} # словарь значений

df_moscow_1["District_cat"] = (
    df_moscow_1["District"].map(district_to_num).fillna(0).astype(int)
)
df_moscow_1.dtypes
# district_to_num
```
